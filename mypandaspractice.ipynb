```python
def grade(score):  # Ensure the parameter name is 'score' (not 'Score')
    if score >= 80:
        return 'A'
    elif score >= 70:
        return 'B'
    elif score >= 60:
        return 'C'
    else:
        return 'D'

# apply the grade function to the 'Average' column
df['Grade'] = df['Average'].apply(grade)
df
```

### Combining DataFrames

```python
df1 = pd.DataFrame({'Name': ['Ann', 'Ben'], 'Score': [80, 75]})
df2 = pd.DataFrame({'Name': ['Cara', 'Dan'], 'Score': [90, 85]})
concat_data = pd.concat([df1,df2], axis=0)
concat_data

```

```python
## combined by columns

df3 = pd.DataFrame({'Math': [80, 70]})
df4 = pd.DataFrame({'English': [90, 85]})

combined_cols = pd.concat([df3, df4], axis=1)
print(combined_cols)
```

### 📌 3. Joins in Pandas (pd.merge)
🔗 Used to combine two DataFrames using SQL-like join operations.
🔹 Types of Joins:
Inner Join: Only matching rows

Left Join: All rows from left, matching from right

Right Join: All rows from right, matching from left

Outer Join: All rows from both

```python
students = pd.DataFrame({
    'StudentID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie']
})

scores = pd.DataFrame({
    'StudentID': [1, 2, 4],
    'Score': [85, 90, 75]
})
```

```python
print(students)
print(scores)
inner_join = pd.merge(students,scores, on='StudentID', how='inner')
print(inner_join)
```

```python
left_join = pd.merge(students,scores, on='StudentID', how='left')
left_join 
```
