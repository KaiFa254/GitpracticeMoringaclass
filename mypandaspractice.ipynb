```python
def grade(score):  # Ensure the parameter name is 'score' (not 'Score')
    if score >= 80:
        return 'A'
    elif score >= 70:
        return 'B'
    elif score >= 60:
        return 'C'
    else:
        return 'D'

# Apply the grade function to the 'Average' column
df['Grade'] = df['Average'].apply(grade)
df
```

```python
grouped_by_age_df = df.groupby('Age')[['Math', 'English', 'Science']].mean()
grouped_by_age_df
```
